```python
import numpy as np
# Funkcja do zamiany 0 na losową liczbę z zakresu 1-8
def replace_zero_with_random(dane):
    # Przechodzimy przez każdą kolumnę
    for column in dane.columns:
        # Sprawdzamy, czy w kolumnie są wartości 0
        dane["days"] = dane["days"].apply(lambda x: np.random.randint(1, 5) if x == 0 else x)
    return dane

# Zamieniamy 0 na losowe liczby
dane = replace_zero_with_random(dane)

# Wyświetlenie zmienionych danych
dane
```

```python
# Zakładam, że dane są już wczytane do df
dane_sorted = dane.sort_values(by=["travelCode", "flightDate"])  # Sortujemy po travelCode i dacie

# Grupowanie danych według podróży
grouped = dane_sorted.groupby(["travelCode", "userCode"], as_index=False).agg({
    "from": "first",          # Pierwsze miejsce to miejsce wylotu
    "to": "first",            # Pierwszy cel podróży
    "flightType": "first",
    "flightPrice": list,      # Lista cen lotów (tam + powrót)
    "flightTime": "sum",
    "flightDistance": "sum",
    "flightAgency": "first",
    "flightDate": list,       # Lista dat lotów (tam + powrót)
    "hotelName": "first",
    "hotelNameLabelled": "first",
    "hotelNameHot": "first",
    "hotelPlace": "first",
    "hotelPlaceLabelled": "first",
    "days": "first",
    "hotelPrice": "first",
    "hotelTotal": "first",    # Noclegi tylko raz
    "userCompany": "first",
    "userName": "first",
    "gender": "first",
    "age": "first",
    "MonthOfTravelling": "first",
    "MonthOfTravellingLabelled": "first"
})

# Rozdzielenie flightDate na dwie osobne kolumny
grouped["flightDateTo"] = grouped["flightDate"].apply(lambda x: x[0] if len(x) > 0 else None)
grouped["flightDateBack"] = grouped["flightDate"].apply(lambda x: x[1] if len(x) > 1 else None)

# Rozdzielenie flightPrice na flightPriceTo i flightPriceBack
grouped["flightPriceTo"] = grouped["flightPrice"].apply(lambda x: x[0] if len(x) > 0 else 0)
grouped["flightPriceBack"] = grouped["flightPrice"].apply(lambda x: x[1] if len(x) > 1 else 0)

# Obliczenie flightPriceTotal (łączny koszt lotów tam i z powrotem)
grouped["flightPriceTotal"] = grouped["flightPriceTo"] + grouped["flightPriceBack"]

# Definiowanie nowych kolumn dla klarowności
grouped["flightFrom"] = grouped["from"]  # Skąd startujemy
grouped["flightTo"] = grouped["to"]  # Dokąd lecimy
grouped["flightBackFrom"] = grouped["to"]  # Skąd wracamy (to samo co 'flightTo')
grouped["flightBackTo"] = grouped["from"]  # Dokąd wracamy (to samo co 'flightFrom')

# Obliczenie poprawionego TotalTravelCost (łączny koszt lotów + hotel)
grouped["TotalTravelCost"] = grouped["flightPriceTotal"] + grouped["hotelTotal"]

# Usuwamy już niepotrzebne kolumny
grouped.drop(columns=["flightDate", "flightPrice", "from", "to"], inplace=True)

# Zmieniamy kolejność kolumn, aby 'flightFrom', 'flightTo', 'flightBackFrom', 'flightBackTo' były na początku
cols = ['userCode', 'flightFrom', 'flightTo', 'flightBackFrom', 'flightBackTo', 'flightType', 'flightPriceTo',
        'flightPriceBack', 'flightPriceTotal', 'flightDateTo', 'flightDateBack', 'hotelName', 'hotelNameLabelled','hotelNameHot', 'hotelPlace','hotelPlaceLabelled', 'days',
        'hotelPrice', 'hotelTotal', 'userCompany', 'userName', 'gender', 'age', 'MonthOfTravelling', 'MonthOfTravellingLabelled', 'TotalTravelCost']
grouped = grouped[cols]

```
